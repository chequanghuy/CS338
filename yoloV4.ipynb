# Tạo file train.txt và valid.txt

```
#Tạo file train.txt, valid.txt theo đoạn code
import os
import numpy as np
#"obj" là tên thư mục chứa cả ảnh và file annotation.
lst_files = os.listdir("data/dataset/")               #đưa tên tất cả file trong folder dataset vào mảng lst_files
lst_images = []                                       #mảng chứa tên ảnh

for file in lst_files:                                #duyệt từng phần tử trong lst_files
  if ".txt" not in file:                              #nếu là ảnh
    lst_images.append(file)                           #đưa vào mảng chứa tên ảnh
    
#Tách 10 ảnh ra làm tập validation  
random_idx = np.random.randint(0, len(lst_images), 100)

#Tạo file train.txt được đặt trong thư mục darknet/data
with open("data/dataset/train.txt","w") as f:
  for idx in range(len(lst_images)):
    if idx not in random_idx:
      f.write("data/dataset/"+lst_images[idx]+"\n")
      
#Tạo file valid.txt được đặt trong thư mục darknet/data
with open("data/dataset/valid.txt","w") as f:
    for idx in random_idx:
      f.write("data/dataset/"+lst_images[idx]+"\n")
```

```
def show(path):
  import cv2
  import matplotlib.pyplot as plt

  image = cv2.imread(path)
  original_width, original_height = image.shape[1], image.shape[0]
  resized_image = cv2.resize(image, (2*original_width, 2*original_height)\
  , interpolation = cv2.INTER_CUBIC)

  resized_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
  plt.figure(figsize=(20,10))
  plt.axis("off")
  plt.imshow(resized_image)
  plt.show()
 
show("predictions.jpg")
```
